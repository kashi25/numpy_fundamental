# concatenating Dataframe ::append them(one over the other)   ::same rows and append them side-by side

```python
# dataframe with same column
df1  = pd.DataFrame({'Fruit':['Apple','Orange','Mango','Kiwi'],
                     'Quantity':['12','34','56','66'],
                     'Box':['A','B','C','D']})
df2 = pd.DataFrame({'Chor':['Puspa','Khadga','Sherbadur','Madav'],
                   'Qualification':['1','2','0','1'],
                   'Deskolagi':['lutnay','jhutvolnay','fata','gadha']}
                  )
```

```python
df2
```

```python
df1
```

```python
df3  = pd.DataFrame({'Fruit':['Guva','Tomato','watermallon'],
                     'Quantity':['11','44','6'],
                     'Box':['G','K','M']})
df3
```

```python
# concate df1 and df3
pd.concat([df1,df3],axis =0)
# axis = indicate concate along the row axis, one over the other
# keep in mind index is merger, i.e index are simpled copied so, we have to reset the index
```

```python
# also use append() function will also give same result
df1.append(df3)
```

```python
# satistics for Nepal footbal
Nepal_footbal = pd.DataFrame({'Nepali internal Team':['manang','JhapaXI','ThreeStar','himalayan'],
                             'Matches played':[14,13,12,40],
                             'Match Won':[10,4,6,22]}
                            )
Nepal_footbal
```

```python
Nepal_footbal.set_index('Nepali internal Team',inplace=True)
Nepal_footbal
```

```python
Nepal_footbal2 = pd.DataFrame({'Nepali internal Team':['Terthum','Jhapa','Khotang','himalayan'],
                             'Matches played':[14,13,12,40],
                             'Match Won':[1,4,6,20]}
                            )
Nepal_footbal2
```

```python
Nepal_footbal2.set_index('Nepali internal Team',inplace=True)
Nepal_footbal2
```

```python
# add the two DFs, it will display add result for match data
add_data = Nepal_footbal + Nepal_footbal2
add_data
```

```python
# there are new teams hence NaN values
# df.add() function to replace Nan values in the two dataFrames
# results index will be sorted alphabetically
add_data = Nepal_footbal.add(Nepal_footbal2, fill_value =0)
add_data
```

```python
# create a new colum - 'Win %'

add_data['Win %'] = add_data['Match Won']/add_data ['Matches played']
add_data
```

```python
# sort the teams with highest number of wins

add_data.sort_values(by =(['Match Won','Win %']),ascending =False)
add_data
```
