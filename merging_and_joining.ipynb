# Data merging and concatenation

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']},
                  index = [0,1,2,3])
```

```python
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                   'B':['B4','B5','B6','B7'],
                   'C':['C4','C5','C6','C7'],
                   'D':['D4','D5','D6','D7']},
                  index = [4,5,6,7])
```

```python
df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                   'B':['B8','B9','B10','B11'],
                   'C':['C8','C9','C10','C11'],
                   'D':['D8','D9','D10','D11']},
                  index = [8,9,10,11])
```

```python
# dataFrame merging
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat(objs=[df1,df2,df3])
```

```python
pd.concat(objs=[df1,df2,df3],axis=1) #axis =1 is along the column
```

## dataframe merging

```python
dfk1 = pd.DataFrame({'k':['R4','R5','R6','R7'],
                   'M':['B0','B1','B2','B3'],
                   'N':['C0','C1','C2','C3'],
                   'O':['D0','D1','D2','D3']},
                  index = [0,1,2,3])
```

```python
dfk2 = pd.DataFrame({'k':['R4','R5','R6','R7'],
                   'S':['S4','s5','s6','s7'],
                   'T':['t4','t5','t6','t7'],
                   'L':['L4','L5','L6','L7']},
                  index = [4,5,6,7])
```

```python
dfk1
```

```python
dfk2
```

```python
pd.merge(left=dfk1,right=dfk2,on='k')
```

```python
dfm1 = pd.DataFrame({'P':['R4','R5','R6','R7'],
                   'M':['B0','B1','B2','B3'],
                   'N':['C0','C1','C2','C3'],
                   'O':['D0','D1','D2','D3']},
                  index = [0,1,2,3])
```

```python
dfm2 = pd.DataFrame({'P':['R4','R5','R6','R7'],
                   'M':['B0','B1','B2','B3'],
                   'E':['C0','C1','C2','C3'],
                   'R':['D0','D1','D2','D3']},
                  index = [4,5,6,7])
```

```python
dfm1
```

```python
dfm2
```

```python
pd.merge(left=dfm1,right=dfm2, on=['P','M'])
```

# dataFrame join

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']},
                   index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[0, 1, 2, 3])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```
