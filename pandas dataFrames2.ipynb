# conditional selection in dataframe

```python
import numpy as np
import pandas as pd
from numpy.random import rand
np.random.seed(101)
```

```python
df = pd.DataFrame(data= rand(5,4), index=['a','b','c','d','e'],columns=['w','x','y','Z'])
```

```python
df
```

```python
# condition selction
df > 0.5
```

```python
b = df>0.5
b
```

```python
df[b]
```

```python
df['w']>0.5
```

```python
d =df['w']>0
d
```

```python
# multiple condition at a time
df
```

```python
df[(df['w']>0) & (df['y']>0.6)]
```

```python
df[(df['w']>0) | (df['y']>0.6)]
```
